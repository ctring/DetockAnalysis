```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 3, figsize=(12, 5), sharex=True)

x = np.arange(1, 100, 20)

y_throughput = [
    100000/(x + 10) + 10000,
    100000/(x + 20) + 10000,
    100000/(x + 30) + 10000,
    100000/(x + 40) + 10000
]

for i, mp in enumerate([0, 50, 100]):
    for y in y_throughput:
        ax = axes[0, i]
        ax.plot(x, y, marker='.')
        ax.grid('on')
        ax.set_title(f'MP Txns = {mp}%')

axes[0, 0].set_ylabel('Throughput (txn/s)')

y_latency = [
    20*np.sqrt(np.sqrt(x)) + 10,
    30*np.sqrt(np.sqrt(x)) + 10,
    40*np.sqrt(np.sqrt(x)) + 10,
    50*np.sqrt(np.sqrt(x)) + 10,
]

for i, mp in enumerate([0, 50, 100]):
    for y in y_latency:
        ax = axes[1, i]
        ax.plot(x, y, marker='.')
        ax.grid('on')
        ax.set_xlabel('% MH Transactions')

axes[1, 0].set_ylabel('Latency (ms)')

fig.suptitle('High contention (HOT = 50)')

plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def zipf_pmf(a, n, x):
    return 1 / (np.array(x) ** a * np.sum(1/np.arange(1, n + 1) ** a))

def plot_zipf(a, n):
    x = np.arange(1, n + 1)
    plt.plot(x, zipf_pmf(a, n, x), marker='o', label=f'a = {a}')
    plt.grid('on')

for a in [0, 0.5, 1.0, 1.5, 2.0, 2.5]:
    plot_zipf(a, 10)

plt.legend()
```
